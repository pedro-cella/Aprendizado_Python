# Obtendo a lista de nomes dos personagens

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd

# Esse código tem como intuito utilizar de técnicas de Web Scrapping para colher dados dos personagens do jogo
# Genshin Impact, utilizando da biblioteca Selenium. 

# Passo 1: Acessar a pagina da web

navegador = webdriver.Chrome()
navegador.get('https://genshin.honeyhunterworld.com/db/char/characters/?lang=EN')
navegador2.get('https://game8.co/games/Genshin-Impact/archives/307054')

# Passo 2: Obter a lista de nomes de todos os personagens

# Declarando variáveis
dados_personagens = {}
lista_personagens = list()
tabela_personagens = list()

# Esse código serve para acessar o site, buscar e salvar o nome dos personagens em uma lista
personagens = navegador.find_elements(By.CLASS_NAME, 'char_sea_cont')
for personagem in personagens:
    lista_personagens.append(personagem.text)

lista_personagens.sort()

print(lista_personagens)

nomes = lista_personagens.copy()

# Esse código serve para transformar o nome dos personagens em minúsculo
# Pedaço necessário para buscar as informações dos mesmos
for i in range(len(nomes)):
    nomes[i] = nomes[i].lower()
```

## Código

```python
# Passo 3: Acessar as paginas dos personagens
nomes = list(set(nomes))

nomes.sort()

# Esse pedaço irá acessar o nome de cada personagem na lista e com isso buscar suas informações
for nome in nomes:
    
    # Passo 4: Obter os as informacoes(Nome, Arma, Elemento e Status do nivel 90) de cada personagem
    dados_personagens["Nome"] = nome.title() # Transforma o nome novamente para a primeira letra maiúscula
    
    # Tratamento do nome de alguns personagens, incluindo nomes com espaço
    if nome == "hu tao" or nome == "yun jin":
        nome = nome.replace(" ", "")
    elif nome == "yanfei":
        nome = "feiyan"
    elif (' ' in nome) == True and nome == "raiden shogun":
        nome = nome.split()
        nome = "shougun"
    elif (' ' in nome) == True:
        nome = nome.split()
        nome = nome[1]
    
    # Acessa a página do personagem
    navegador.get(f"https://genshin.honeyhunterworld.com/db/char/{nome}/?lang=EN")
    
    # Busca a arma do personagem e salva no dicionário
    arma = navegador.find_element(By.XPATH, '//*[@id="live_data"]/table[1]/tbody/tr[5]/td[2]/a').text
    dados_personagens["Arma"] = arma
    
    # Busca o elemento do personagem comparando o nome do elemento com o título existente na imagem e 
    # salva no dicionário
    imagem = navegador.find_element(By.XPATH, '//*[@id="live_data"]/table[1]/tbody/tr[6]/td[2]/img').get_attribute('data-src')
    if 'cryo' in imagem:
        elemento = 'Cryo'
    elif 'pyro' in imagem:
        elemento = 'Pyro'
    elif 'geo' in imagem:
        elemento = 'Geo'
    elif 'electro' in imagem:
        elemento = 'Electro'
    elif 'hydro' in imagem:
        elemento = 'Hydro'
    elif 'anemo' in imagem:
        elemento = 'Anemo'
    dados_personagens["Elemento"] = elemento
    
    # Busca o HP BASE, ATAQUE BASE, DEFESA BASE, TAXA CRÍTICA e DANO CRÌTICO do personagem no nivel 90
    # e salva no dicionário
    base_hp = navegador.find_element(By.XPATH, '//*[@id="live_data"]/div[1]/table/tbody/tr[15]/td[2]').text
    base_atk = navegador.find_element(By.XPATH, '//*[@id="live_data"]/div[1]/table/tbody/tr[15]/td[3]').text
    base_defense = navegador.find_element(By.XPATH, '//*[@id="live_data"]/div[1]/table/tbody/tr[15]/td[4]').text
    crit_rate = navegador.find_element(By.XPATH, '//*[@id="live_data"]/div[1]/table/tbody/tr[15]/td[6]').text
    crit_dmg = navegador.find_element(By.XPATH, '//*[@id="live_data"]/div[1]/table/tbody/tr[15]/td[7]').text
    dados_personagens["Nivel"] = 90
    dados_personagens["HP Base"] = base_hp
    dados_personagens["ATK Base"] = base_atk
    dados_personagens["Defesa Base"] = base_defense
    dados_personagens["Taxa Critica"] = crit_rate
    dados_personagens["Dano Critico"] = crit_dmg
    
    # Busca a raridade do personagem contando quantos elementos existem na descrição da raridade
    # no caso o site utiliza de imagens de estrela, logo contando quantas tem é possível diferenciar
    # um personagem 4 estrelas e um 5 estrelas
    tag_pai = navegador.find_element(By.XPATH, '//*[@id="live_data"]/table[1]/tbody/tr[4]/td[2]')
    raridade = len(tag_pai.find_elements(By.TAG_NAME, 'div'))
    dados_personagens["Raridade"] = raridade
    tabela_personagens.append(dados_personagens.copy())
    
print(tabela_personagens)
```

```python
# Gera um Data Frame com todos os dados salvos no dicionário
dataset = pd.DataFrame(tabela_personagens)
dataset
```
