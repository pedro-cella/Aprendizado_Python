```python
# Importano bibliotecas
import pandas as pd
import matplotlib as mt
```

```python
# Carregando dados do csv
df1 = pd.read_csv('./Dataset/Data_Collection_Chars.csv')
df2 = pd.read_csv('./Dataset/Data_Collection_Stats.csv')
```

```python
df1
```

```python
df2
```

```python
# Juntando dados em uma unica tabela
df3 = pd.merge(df1, df2, on="Personagem")
df3
```

# Exploratory Data Analysis

## Dados duplicados

```python
df3_duplicado = df3.Personagem.duplicated()
df3_duplicado
```

## Visualizando dados duplicados

```python
df3.loc[df3.duplicated(), :]
```

## Eliminando dados duplicados

```python
novo_df3 = df3.drop_duplicates(keep='first')
novo_df3
```

## Quantidade de personagens por elemento

## Com dados duplicados

```python
elementos = df3['Elemento'].value_counts()
elementos
```

## Sem dados duplicados

```python
elementos = novo_df3['Elemento'].value_counts()
elementos
```

```python
# Salvando tabela em um csv
novo_df3.to_csv('./Dataset/Data_Collection_Chars_Data.csv', index=False, encoding = 'utf-8-sig')
```
