```python
# Importando bilbiotecas
from urllib.request import urlopen
from bs4 import BeautifulSoup
import wordninja
import re
import pandas as pd
```

```python
# acessando URL da pagina
url = 'https://gamewith.net/genshin-impact/article/show/22357'

response = urlopen(url)

# Obtendo HTML da pagina e limpando infos
html = response.read().decode('utf-8')
html.split()
def trata_html(input):
    return " ".join(input.split()).replace('> <', '><')
html = trata_html(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# Obtendo div dos dados que serao retirados
tabela = soup.find('div', class_='genshin_cha')
dados = tabela.findAll('tr')

# Declarando variaveis
dados_personagem = {}
personagens = []

# Iterando lista de informacoes retiradas da pagina(Nome, Elemento e Arma)
for i in range(1, len(dados)):
    infos = str(dados[i].get_text()).replace(" ", "")
    infos_separadas = re.findall('[a-zA-Z][^A-Z]*', infos)
#     print(infos_separadas)
    if infos_separadas[0] == "Hu" or infos_separadas[0] == "Raiden":
        nome = infos_separadas[0] + " " + infos_separadas[1]
        dados_personagem['nome'] = nome
        dados_personagem['elemento'] = infos_separadas[2]
        dados_personagem['arma'] = infos_separadas[3]
    elif infos_separadas[0] == "Anemo" or infos_separadas[0] == "Geo" or infos_separadas[0] == "Electro":
        dados_personagem['nome'] = infos_separadas[1]
        dados_personagem['elemento'] = infos_separadas[0]
        dados_personagem['arma'] = infos_separadas[3]
    else:
        dados_personagem['nome'] = infos_separadas[0]
        dados_personagem['elemento'] = infos_separadas[1]
        dados_personagem['arma'] = infos_separadas[2]
        
    personagens.append(dados_personagem.copy())
    
# Criando DataFrame e gerando CSV
dataset = pd.DataFrame(personagens)
dataset = dataset.sort_values(by='nome')
dataset.to_csv('./Dataset/personagens.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
```
