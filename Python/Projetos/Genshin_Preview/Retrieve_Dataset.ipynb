```python
# Importando bibliotecas
from urllib.request import urlopen, urlretrieve, Request
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
```

# Criando uma lista com os nome dos personagens

```python
# Fazendo requisicao para  a URL
url = 'https://genshin.honeyhunterworld.com/db/char/characters/?lang=EN'
headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:95.0) Gecko/20100101 Firefox/95.0'}

# Declarando variaveis
lista = []

# Realizando um tratamento de erro caso a requisicao nao funcione
try:
    # Obtendo HTML
    req = Request(url, headers = headers)
    response = urlopen(req)
    html = response.read().decode('utf-8')
    
    # Tratamento de HTML -> String
    def trata_html(input):
        return " ".join(input.split()).replace('> <', '><')
    
    html = trata_html(html)
    
    # Utilizando do BeautifulSoup para obter o nome dos personagens
    soup = BeautifulSoup(html, 'html.parser')
    dados = soup.find('div', {'class': 'wrappercont wrappercont_char'}).findAll('div', class_='char_sea_cont')
    
    # Coletando nome dos personagens
    for dado in dados:
        lista.append(dado.find('span', {'class': 'sea_charname'}).get_text())
        
    print(lista)
    
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e:
    print(e.reason)
```
